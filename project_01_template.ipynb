## 4. Summarizing Data

## What is the relationship between US Sales(Millions) and Console in this dataset?

```python
sns.boxplot(x="Console", y= "US Sales (millions)", data=df)
```

```python
df["US Sales (millions)"].describe()
```

## The boxplot of US Sales (millions) by Console shows how sales are distributed across different gaming platforms. From this visualization, we can compare consoles to see which platforms tend to have higher or lower US sales, helping answer the research question: What is the relationship between US Sales and Console in this dataset?

```python
summary_stats = df.groupby('Console')['US Sales (millions)'].agg(['count', 'mean', 'median', 'min', 'max', 'std'])
print(summary_stats)
```
